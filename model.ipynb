```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataset_diabetes/diabetic_data.csv")
list(df)
features = ['age', 'race', 'gender', 'num_lab_procedures']
```

```python
#num of readmitted with age
unique_df = df.copy()
unique_df = unique_df.drop_duplicates(subset = "patient_nbr", keep= "first")
print(len(unique_df))
#num of readmitted w/o age
readmitted = unique_df[unique_df['readmitted'] != 'NO']
set(list(readmitted['readmitted']))

ages = set(list(readmitted['age']))
ages_prob = {}
for elem in ages:
    ages_prob[elem] = len(readmitted[readmitted['age']==elem])/len(readmitted)
print(ages_prob)
print(sum(ages_prob.values()))

#for all data
ages_prob_all = {}
for elem in ages:
    ages_prob_all[elem] = len(unique_df[unique_df['age']==elem])/len(unique_df)
print(ages_prob_all)
print(sum(ages_prob_all.values()))


prob_disease = len(readmitted)/len(unique_df)
for patient_age in ages:
    prob_disease_given_patient_age = prob_disease * ages_prob[patient_age] / ages_prob_all[patient_age]
    print("for age range %s, prob is %f" %(patient_age, prob_disease_given_patient_age))
```

```python
races = set(list(readmitted['race']))
races_prob = {}
for elem in races:
    races_prob[elem] = len(readmitted[readmitted['race']==elem])/len(readmitted)
print(races_prob)
print(sum(races_prob.values()))

#for all data
races_prob_all = {}
for elem in races:
    races_prob_all[elem] = len(unique_df[unique_df['race']==elem])/len(unique_df)
print(races_prob_all)
print(sum(races_prob_all.values()))


prob_disease = len(readmitted)/len(unique_df)
for patient_race in races:
    prob_disease_given_patient_race = prob_disease * races_prob[patient_race] / races_prob_all[patient_race]
    print("for race range %s, prob is %f" %(patient_race, prob_disease_given_patient_race))
```

```python
num_labs = set(list(readmitted['num_lab_procedures']))
num_labs_prob = {}
for elem in num_labs:
    num_labs_prob[elem] = len(readmitted[readmitted['num_lab_procedures']==elem])/len(readmitted)
print(sum(num_labs_prob.values()))

#for all data
num_labs_prob_all = {}
for elem in num_labs:
    num_labs_prob_all[elem] = len(unique_df[unique_df['num_lab_procedures']==elem])/len(unique_df)
print(sum(num_labs_prob_all.values()))

prob_disease = len(readmitted)/len(unique_df)
for patient_num_lab in num_labs:
    prob_disease_given_patient_num_lab = prob_disease * num_labs_prob[patient_num_lab] / num_labs_prob_all[patient_num_lab]
    print("for num_labs range %s, prob is %f" %(patient_num_lab, prob_disease_given_patient_num_lab))
    print("Number of patients with %s number of lab procedures: %f" %(patient_num_lab, len(unique_df[unique_df['num_lab_procedures']==patient_num_lab])))
```
